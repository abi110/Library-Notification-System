1.
Write code to produce a report of the frequency with which books have been loaned in reverse order
(the least frequent first). The report should indicate the number of times that books were loaned in
2019, along with their titles and author. The latter are required as there are some texts (classics in
some way) which will be retained even if they have proved unpopular.

```python
# import libraries
import csv   
import pandas as pd
import numpy as np

# opening the CSV file   
with open('books.csv', mode ='r') as file:   
    # reading the CSV file   
    reader = csv.reader(file)
    
# add headers to data
headerList = ['Number', 'Title', 'Author', 'Genre', 'SubGenre', 'Publisher']
# using pandas to read the data
books = pd.read_csv('books.csv')
# using pandas to write headers to books.csv
books.to_csv('books.csv',header=headerList, index=False)

# opening the CSV file   
with open('bookloans.csv', mode ='r') as file:   
    # reading the CSV file   
    reader = csv.reader(file)
    # writing to the csv file
    writer = csv.writer(file, delimiter=',')
    
# adding headers
headerList = ['book-number', 'member-number', 'date-of-loan', 'date-of-return']
# using pandas to read the data
bookloans = pd.read_csv('bookloans.csv')
# using pandas to write headers to books.csv
bookloans.to_csv('bookloans.csv',header=headerList, index=False)

# cleaning data so only loans from within 2019 remain
# defining cdf (cleaned data frame) as shorthand for bookloans
cdf = bookloans
# loop writen to clean data so only 2019 remains,
# uses 43465 epoch for 01/01/2019 and 43830 epoch for 31/12/2019,
# loop also drops books with booknumber(n) over 120
x=0
while x < 2208:
    for row in cdf:
        l = cdf.loc[x,'date-of-loan']
        r = cdf.loc[x,'date-of-return']
        n = cdf.loc[x,'book-number']
        if l < 43465:
            cdf.drop(x,axis=0,inplace=True)
        elif r > 43830:
            cdf.drop(x,axis=0,inplace=True)
        elif l > 43830:
            cdf.drop(x,axis=0,inplace=True)
        elif n > 120:
            cdf.drop(x,axis=0,inplace=True)
        x = x+1

# list created of book numbers only to allow for counting of the frequency each book was loaned
book_number_only = list(cdf['book-number'])
# function defined to count frequency of each book loaned and return as a dictionary
def CountFreq(book_number_only):
    freq = {}
    for items in book_number_only:
        freq[items] = book_number_only.count(items)
    return(freq)

# sorting book loans in reverse order (least frequent first)
dict1 = CountFreq(book_number_only)
# defined as sorted_dict
sorted_dict1={k: v for k, v in sorted(dict1.items(), key=lambda item: item[1])}
# separating sorted_dict into separate lists of keys and values
key_list1 = list(sorted_dict1.keys())
val_list1 = list(sorted_dict1.values())

# defining lookup function to return 
def lookup():
    x=0
    while x < len(books):
        k = key_list1[x]
        f = val_list1[x]
        t = books.loc[(k-1),'Title']
        a = books.loc[(k-1),'Author']
        print(k, f, t, a)
        x = x+1

# Report Printed
print("Report indicating the book number, frequency that book was loaned in 2019, along with the book's titles and author.")
print("")
lookup()
```

3. 
Write code to produce a report of the average length of time that a user has a book on loan and the 
proportion of users who have returned books late, together with the average late period. 

```python
# List of only date of loan created
date_loaned = list(cdf['date-of-loan'])
# List of only date returned created
date_returned = list(cdf['date-of-return'])

# Function returns defined to create list of how long books were loaned for
def returns():
    list = []
    x=0
    while x < len(date_returned):
        # finds days between date loaned and date returned
        r = date_returned[x]-date_loaned[x]
        # a 0 indicates that the book was not returned
        if r > 0:
            list.append(r)
            # a score higher than 0 means that the book was returned so value added to functions list
        else:
            # a 0 indicates that the book was not returned so is passed and is not added to the list
            if r < 0:
                pass
        x = x+1          
    return list

# Function late defined to create list of books returned late
def late():
    list = []
    x=0
    while x < len(date_returned):
        r = date_returned[x]-date_loaned[x]
        # 14 days is the library's standard loan period so books that were loaned for longer than 14 days can be considered late
        # loans of more than 14 days added to the list created by the function
        if r > 14:
            list.append(r)
        else:
            pass
        x = x+1          
    return list

# Calculation for average length of loan
avg_length = sum(returns()) / len(returns())
# Calculation for percentage of loans that were late
percentage_late = (len(late()) / len(returns()))*100
# Calculation for average late period
avg_late = sum(late())/len(late())

# Report Printing
print(f"The average length of time that a member has a book is {int(avg_length)} days")
print(f"The percentage of book loans that were returned late is {int(percentage_late)}%")
print(f"The average length of time that those who returned late had the book for is {int(avg_late)} days")
print(f"The average period of time that a book was late by is {int(avg_late) - 14} days")
```
